## SVM

В этом задании Вы увидите, насколько хорошо kernel trick  справляется с нелинейной зависимостью между независимыми переменными и целевой. Также можно убедиться, что SVM хорош для работы с пропущенными данными.

Пусть у нас есть выборка из зелёных и красных точек. Красные точки образуют круг внутри зелёных.

Представим, что у нас отсутствует 1/3 данных. Мы хотим восстановить их.

Построим модель на имеющихся данных. В качестве моделей попробуем логистическую регрессию, решающее дерево и SVM. Посмотрим на их предсказания для отсутствующих данных.

1) Создайте массивы координат x: от 0 до 15 с шагом 0.1,  y: от 0 до 10 с шагом 0.1 

2) Из векторов x,y создайте координатную сетку, используя *numpy.meshgrid* [https://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html]

3) Получившиеся координаты запишите в Dataframe

4) Визуализируйте данные

  4.1 Добавьте в таблицу атрибут , равный $(x-7.5)^2 + (y-5)^2$. Он определит наш круг
  
  4.2 Добавьте в таблицу атрибут цвета. Если атрибут из пункта 4.1 < 15 (т.е. точки лежат внутри круга, радиус которого $\sqrt{15}$), то цвет=1 (обозначим красный единицей), иначе - 0(обозначим так зелёный)
  
  **NB Атрибут цвета - целевая метка класса, т.е. именно это значение должно предсказываться классификатором.**
  
  **Признаки,на которых обучаем модели, - только координаты (x,y)**
  
  4.3 Визуализируйте данные, используя *pyplot.scatter*. Для точек, у которых атрибут цвета =1, используйте красный цвет. Для точек с атрибутом 0 - зелёный.

5) В качестве обучающей выборки возьмите точки, у которых $x<10$, оставшуюся часть - в качестве тестовой выборки. Как раз получим 1/3 данных на тест.

6) Обучите следующие модели и получите предсказания для тестовой выборки: 

**svm.SVC(degree=0.5), **

7) Для каждого классификатора отрисуйте и обучающую, и тестувую выборки в соответствии с предсказанным цветом.

8) Для наглядности нарисуйте на каждом графике вертикальную линию ($x=10$),отделяющую тестовую выборку от обучающей
